```python
import scipy
from scipy import io

DATA =  scipy.io.loadmat('data/einstein_mask.mat')
backgroundMask = DATA['mask']>0
```

```python
plt.imshow(backgroundMask,cmap='gray')
```

```python
image = imread('data/einstein.tif')
imageGrayBackground = np.zeros(image.shape)
imageGrayBackground[:,:] = image
imageGrayBackground[backgroundMask] = 50
```

```python
fig = figure(figsize=(12,4), dpi=200)

fig.add_subplot(1,3,1)
img = imshow(image, cmap='gray')
img.axes.get_xaxis().set_visible(False)
img.axes.get_yaxis().set_visible(False)

fig.add_subplot(1,3,2)
img = imshow(backgroundMask, cmap='gray')
img.axes.get_xaxis().set_visible(False)
img.axes.get_yaxis().set_visible(False)

fig.add_subplot(1,3,3)
img = imshow(imageGrayBackground, cmap='gray')
img.axes.get_xaxis().set_visible(False)
img.axes.get_yaxis().set_visible(False)
```
